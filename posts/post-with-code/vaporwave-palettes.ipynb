```python

# This could be named vaporpalettes, maybe be a class (to do something like vaporpalettes.cool for a palette), and be a package to import it.  
palettes = {
    "vaporwave": ["#94D0FF", "#8795E8", "#966bff", "#AD8CFF", "#C774E8", "#c774a9", "#FF6AD5", "#ff6a8b", "#ff8b8b", "#ffa58b", "#ffde8b", "#cdde8b", "#8bde8b", "#20de8b"],
    "cool": ["#FF6AD5", "#C774E8", "#AD8CFF", "#8795E8", "#94D0FF"],
    "crystal_pepsi": ["#FFCCFF", "#F1DAFF", "#E3E8FF", "#CCFFFF"],
    "mallsoft": ["#fbcff3", "#f7c0bb", "#acd0f4", "#8690ff", "#30bfdd", "#7fd4c1"],
    "jazzcup": ["#392682", "#7a3a9a", "#3f86bc", "#28ada8", "#83dde0"],
    "sunset": ["#661246", "#ae1357", "#f9247e", "#d7509f", "#f9897b"],
    "macplus": ["#1b4247", "#09979b", "#75d8d5", "#ffc0cb", "#fe7f9d", "#65323e"],
    "seapunk": ["#532e57", "#a997ab", "#7ec488", "#569874", "#296656"],
    "avanti": ["#FB4142", "#94376C", "#CE75AD", "#76BDCF", "#9DCFF0"],
    "neon_night": ["#FF61A6", "#D783FF", "#7161EF", "#3A4E9B", "#2D3045"],
    "pastel_dream": ["#FFB1C1", "#FFCF56", "#8DFFB9", "#80A1C1", "#D4A5A5"],
    "retro_future": ["#FF6B6B", "#00A3E1", "#FFB3E1", "#C4C4A5", "#2E2E2E"],
    "digital_ocean": ["#05C1F0", "#0B61A4", "#5A009D", "#B80058", "#FF0000"],
    "synthwave": ["#FF0055", "#FB4D3D", "#FED304", "#00E156", "#009FB7"],
    "chillwave": ["#A1FFFE", "#719D89", "#FFC3A7", "#D0A3BF", "#81559B"],
    "dreamscape": ["#DAFFEF", "#B2B2FF", "#FFC3C2", "#FAA7B8", "#D4A5E3"],
    "cyberpunk": ["#F6019D", "#7200DA", "#01C0C8", "#FF570A", "#FFAA00"],
    # NEW, MAYBE REMOVE
    "dreamy_sunset": ["#FFABAB", "#FFC3FF", "#D4A5FF", "#B0E0E6", "#94DFFF"],
    "nostalgia_neon": ["#FF4EA6", "#00B2FF", "#AB83A1", "#23007B", "#090909"],
    "ocean_breeze": ["#1BE7FF", "#6EEB83", "#E4FF1A", "#E8AA14", "#FF5714"],
    "cyber_paradise": ["#FF00E7", "#6A00FF", "#006AFF", "#00FFB3", "#DFFF00"],
    "retro_beach": ["#FFC3A0", "#FF61A6", "#AB83A1", "#23007B", "#533D7F"],
    "future_city": ["#FE53BB", "#F5D300", "#00DBE1", "#74F2CE", "#ACB6E5"],
    "tropical_twilight": ["#B1E5FB", "#FF61EA", "#D65EF5", "#8458B3", "#4B0082"],
    "neon_nature": ["#FF00CC", "#3333FF", "#66FF66", "#FF3300", "#FFFF00"],
}

avant_garde_palettes = {
    "glitch_art": ["#FF00FF", "#00FF00", "#0000FF", "#FF0000", "#00FFFF"],
    "futurist_fantasy": ["#C1FF00", "#FF0090", "#9400D3", "#4B0082", "#00008B"],
    "abstract_expression": ["#C51162", "#00796B", "#FFB300", "#6200EA", "#795548"],
    "urban_decay": ["#BCAAA4", "#D32F2F", "#0288D1", "#7CB342", "#FBC02D"],
    "surreal_sunset": ["#8E44AD", "#E74C3C", "#3498DB", "#F39C12", "#2ECC71"],
    "cubist_cool": ["#D81B60", "#1E88E5", "#43A047", "#FDD835", "#FB8C00"],
    "dada_dive": ["#AD1457", "#1976D2", "#388E3C", "#F9A825", "#F4511E"],
    "bauhaus_blast": ["#C2185B", "#1976D2", "#00796B", "#AFB42B", "#E64A19"],
    "metaphysical_mind": ["#880E4F", "#1565C0", "#2E7D32", "#F9A825", "#D84315"],
    "neoplasticism": ["#D50000", "#304FFE", "#00C853", "#FFD600", "#000000"],
}

# palettes.update(avant_garde_palettes)


import matplotlib.pyplot as plt

def display_palettes(palettes):
    n = len(palettes)
    rows = int(n**0.5)
    cols = int(n / rows) + (n % rows > 0)

    fig, axes = plt.subplots(rows, cols, figsize=(cols * 2.5, rows * 2))
    for ax in axes.flatten():
        ax.axis('off')

    for idx, (name, colors) in enumerate(palettes.items()):
        row = idx // cols
        col = idx % cols
        ax = axes[row, col]
        ax.axis('on')
        for j, c in enumerate(colors):
            ax.hlines(j, 0, 1, colors=c, linewidth=15)
        ax.set_ylim(-1, len(colors))
        ax.set_title(name)
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['left'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_ticks([])

    plt.tight_layout()
    plt.show()

display_palettes(palettes)
```
